```python
import matplotlib.pyplot as plt
import japanize_matplotlib
import sympy as sp
from IPython.display import display, Math
import time

# ガウス和を公式で計算する関数
def gauss_sum_formula(n):
    return n * (n + 1) // 2

# ガウス和を逐次計算で求める関数
def iterative_gauss_sum(n):
    total = 0
    terms = []
    for i in range(1, n + 1):
        total += i
        terms.append(str(i))
    return total, " + ".join(terms)

# nを指定
N = 10

# 公式を使った結果
start_time = time.time()
formula_result = gauss_sum_formula(N)
time_formula = time.time() - start_time

# 逐次計算を使った結果
start_time = time.time()
iterative_result, sum_expression = iterative_gauss_sum(N)
time_loop = time.time() - start_time


# SymPyを使って公式を数式として表示
n = sp.Symbol('n')


# LaTeXで数式を表示
display(Math(fr"n = {N}"))
display(Math(fr"S = {sum_expression}"))
# 結果を表示
print(f"逐次計算を使った結果: {iterative_result}\n")

# LaTeXで数式を表示
display(Math(r"S = \frac{n(n + 1)}{2}"))
# 結果を表示
print(f"公式を使った結果: {formula_result}\n")
# 逐次計算を使った足し算の形式で表示


# 計算時間を棒グラフで表示
methods = [r"$n(n+1)/2$", r"$\sum_{k=1}^{n}k$"]
times = [time_formula, time_loop]

plt.figure(figsize=(8, 6))
plt.bar(methods, times, color=['blue', 'yellow'])
plt.ylabel('時間 (秒)')
plt.title('ガウス和の計算時間の比較')
plt.show()
```
