```python
import matplotlib.pyplot as plt
import time
import sympy as sp

def tenth_power_sum_formula(n):
    """
    
    
    Args:
    n (int):  n
    
    Returns:
    int: 
    """
    x = sp.symbols('x')
    sum_expr = sp.summation(x**10, (x, 1, n))
    return sum_expr

def tenth_power_sum_loop(n):
    """
    
    
    Args:
    n (int):  n
    
    Returns:
    int: 
    """
    sum_result = 0
    for i in range(1, n + 1):
        sum_result += i**10
    return sum_result

# n 
n_values = range(1, 101)  # 1100

# 
tenth_power_sums_formula = [tenth_power_sum_formula(n) for n in n_values]
tenth_power_sums_loop = [tenth_power_sum_loop(n) for n in n_values]

# 
print("n\tFormula\t\tLoop")
for i in range(len(n_values)):
    print(f"{n_values[i]}\t{tenth_power_sums_formula[i]}\t{tenth_power_sums_loop[i]}")

# 
plt.figure(figsize=(10, 6))

# 
plt.plot(n_values, tenth_power_sums_formula, marker='o', linestyle='-', color='b', label='Formula')

# 
plt.plot(n_values, tenth_power_sums_loop, marker='x', linestyle='--', color='r', label='Loop')

plt.title('Sum of Tenth Powers for n=1 to 100')
plt.xlabel('n')
plt.ylabel('Sum')
plt.legend()
plt.grid(True)
plt.show()

# 
n_value = 100000

# 
start_time = time.time()
result_formula = tenth_power_sum_formula(n_value)
time_formula = time.time() - start_time

# 
start_time = time.time()
result_loop = tenth_power_sum_loop(n_value)
time_loop = time.time() - start_time

# 
print(f"Time taken using formula: {time_formula:.6f} seconds")
print(f"Time taken using loop: {time_loop:.6f} seconds")

# 
methods = ['Formula', 'Loop']
times = [time_formula, time_loop]

plt.figure(figsize=(8, 6))
plt.bar(methods, times, color=['blue', 'red'])
plt.ylabel('Time (seconds)')
plt.title('Comparison of Calculation Time for Sum of Tenth Powers')
plt.show()
```
