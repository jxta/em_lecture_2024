```python
import matplotlib.pyplot as plt
import time

def square_sum_formula(n):
    """
    二乗和を公式を使って計算する関数。
    
    Args:
    n (int): 上限の整数値 n。
    
    Returns:
    int: 二乗和の計算結果。
    """
    return n * (n + 1) * (2*n + 1) // 6

def square_sum_loop(n):
    """
    二乗和をループを使って計算する関数。
    
    Args:
    n (int): 上限の整数値 n。
    
    Returns:
    int: 二乗和の計算結果。
    """
    sum_result = 0
    for i in range(1, n + 1):
        sum_result += i**2
    return sum_result

n_value = 10
# n の値を設定
n_values = range(1, n_value+1)  # 1から100までの値

# 二乗和を計算
square_sums_formula = [square_sum_formula(n) for n in n_values]
square_sums_loop = [square_sum_loop(n) for n in n_values]


# 数値での配列表示
print("n\tFormula\t\tLoop")
for i in range(len(n_values)):
    print(f"{n_values[i]}\t{square_sums_formula[i]}\t{square_sums_loop[i]}")
    
# グラフを作成
plt.figure(figsize=(10, 6))

# 公式を使った二乗和をプロット
plt.plot(n_values, square_sums_formula, marker='o', linestyle='-', color='b', label='Formula')

# ループを使った二乗和をプロット
plt.plot(n_values, square_sums_loop, marker='x', linestyle='--', color='r', label='Loop')

n_value_str = str(n_value)
plt.title('Sum of Squares for n=1 to '+ n_value_str)
plt.xlabel('n')
plt.ylabel('Sum')
plt.legend()
plt.grid(True)
plt.show()

# 計算時間の比較


# 公式を使って計算
start_time = time.time()
result_formula = square_sum_formula(n_value)
time_formula = time.time() - start_time

# ループを使って計算
start_time = time.time()
result_loop = square_sum_loop(n_value)
time_loop = time.time() - start_time

# 計算時間を表示
print(f"Time taken using formula: {time_formula} seconds")
print(f"Time taken using loop: {time_loop} seconds")
```
