```python
import numpy as np
import matplotlib.pyplot as plt

def generate_pascals_triangle(n):
    triangle = np.zeros((n, n), dtype=int)
    for i in range(n):
        triangle[i, 0] = 1
        for j in range(1, i + 1):
            triangle[i, j] = triangle[i - 1, j - 1] + triangle[i - 1, j]
    return triangle

def plot_pascals_triangle(triangle):
    n = len(triangle)
    fig, ax = plt.subplots(figsize=(40, 20))
    ax.set_xlim(-n, n)
    ax.set_ylim(-1, n)
    ax.axis('off')
    
    for i in range(n):
        for j in range(i + 1):
            if triangle[i, j] != 0:
                x = j - i / 2
                y = n - 1 - i
                ax.text(x, y, str(triangle[i, j]), ha='center', va='center', fontsize=16)
    
    plt.show()

# パスカルの三角形の行数
n = 10
triangle = generate_pascals_triangle(n)
plot_pascals_triangle(triangle)
```
