```python
import matplotlib.pyplot as plt
import numpy as np

# Barnsley Fern's IFS parameters
# Each row corresponds to a transformation: [a, b, c, d, e, f, probability]
transformations = np.array([
    [0.0, 0.0, 0.0, 0.16, 0.0, 0.0, 0.01],
    [0.85, 0.04, -0.04, 0.85, 0.0, 1.6, 0.85],
    [0.2, -0.26, 0.23, 0.22, 0.0, 1.6, 0.07],
    [-0.15, 0.28, 0.26, 0.24, 0.0, 0.44, 0.07]
])

# Function to apply a transformation
def apply_transformation(x, y, transformation):
    a, b, c, d, e, f, _ = transformation
    new_x = a * x + b * y + e
    new_y = c * x + d * y + f
    return new_x, new_y

# Function to choose a transformation based on probabilities
def choose_transformation(transformations):
    probabilities = transformations[:, 6]
    return transformations[np.random.choice(range(len(transformations)), p=probabilities)]

# Number of points to generate
n_points = 100000

# Starting point
x, y = 0, 0

# Lists to store the points
x_points = []
y_points = []

# Generate the points
for _ in range(n_points):
    transformation = choose_transformation(transformations)
    x, y = apply_transformation(x, y, transformation)
    x_points.append(x)
    y_points.append(y)

# Plot the points
plt.figure(figsize=(10, 10))
plt.scatter(x_points, y_points, s=0.1, color='green')
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Tree FractalのIFSパラメータ
transformations_tree = np.array([
    [0.14, 0.01, 0.0, 0.51, -0.08, -1.31, 0.1],
    [0.43, 0.52, -0.45, 0.5, 1.49, -0.75, 0.35],
    [0.45, -0.49, 0.47, 0.47, -1.62, -0.74, 0.35],
    [0.49, 0.0, 0.0, 0.51, 0.02, 1.62, 0.2]
])

# 変換を適用する関数
def apply_transformation(x, y, transformation):
    a, b, c, d, e, f, _ = transformation
    new_x = a * x + b * y + e
    new_y = c * x + d * y + f
    return new_x, new_y

# 確率に基づいて変換を選択する関数
def choose_transformation(transformations):
    probabilities = transformations[:, 6]
    return transformations[np.random.choice(range(len(transformations)), p=probabilities)]

# 生成するポイントの数
n_points = 100000

# 初期点
x, y = 0, 0

# ポイントを保存するリスト
x_points = []
y_points = []

# ポイントを生成
for _ in range(n_points):
    transformation = choose_transformation(transformations_tree)
    x, y = apply_transformation(x, y, transformation)
    x_points.append(x)
    y_points.append(y)

# ポイントをプロット
plt.figure(figsize=(10, 10))
plt.scatter(x_points, y_points, s=0.1, color='green')
plt.axis('off')
plt.show()
```
